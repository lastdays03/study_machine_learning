### 1. Numpy: ë°°ì—´(Array) ë‹¤ë£¨ê¸°
1ë¶€í„° 10ê¹Œì§€ì˜ ìˆ«ìê°€ ë“¤ì–´ìˆëŠ” ë°°ì—´ì„ ë§Œë“¤ê³ , ë²¡í„° ì—°ì‚°(ëª¨ë“  ìˆ«ìì— 10 ë”í•˜ê¸°)ì„ ìˆ˜í–‰í•´ ë´…ë‹ˆë‹¤.

```python
import numpy as np

# 1~10 ë°°ì—´ ìƒì„±
np_array = np.array(range(1, 11))

# ëª¨ë“  ìš”ì†Œì— 10 ë”í•˜ê¸° (Broadcasting)
print(f"Original: {np_array}")
print(f"Plus 10:  {np_array + 10}")
```

### 2. Pandas: ë°ì´í„°í”„ë ˆì„(DataFrame) ìƒì„± ë° ì¡°íšŒ
ê°€ìƒì˜ ì–´ë²¤ì ¸ìŠ¤ ë°ì´í„°ë¥¼ DataFrameìœ¼ë¡œ ë§Œë“¤ê³  ì •ë³´ë¥¼ ì¡°íšŒí•©ë‹ˆë‹¤.

```python
import pandas as pd

# ë°ì´í„° ì§ì ‘ ìƒì„±
data = {
    'Name': ['Iron Man', 'Captain', 'Thor', 'Hulk'],
    'Age': [48, 100, 1500, 49],
    'Team': ['Tech', 'Leader', 'God', 'Power']
}
df = pd.DataFrame(data)

# ë°ì´í„° í™•ì¸
print("--- Head (Top 2) ---")
print(df.head(2))

print("\n--- Info ---")
print(df.info())
```

### 3. ë°ì´í„° í•„í„°ë§ ë° ì¡°ì‘
*   **í•„í„°ë§**: ë‚˜ì´ê°€ 100ì‚´ ì´ìƒì¸ ì˜ì›…ë§Œ ì¶”ì¶œ
*   **ì—´ ì¶”ê°€**: 'Power'ë¼ëŠ” ìƒˆë¡œìš´ ì»¬ëŸ¼ ì¶”ê°€

```python
# ë‚˜ì´ê°€ 100ì‚´ ì´ìƒì¸ ë°ì´í„° í•„í„°ë§
old_heroes = df[df['Age'] >= 100]
print("--- Heroes over 100 years old ---")
print(old_heroes)

# ìƒˆë¡œìš´ ì»¬ëŸ¼ ì¶”ê°€
df['Power'] = 100
print("\n--- After adding Power column ---")
print(df)
```
