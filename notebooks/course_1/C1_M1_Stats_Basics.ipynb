### 1. Îç∞Ïù¥ÌÑ∞ ÏÉùÏÑ± (Random Generation)
ÌèâÍ∑†Ïù¥ 0Ïù¥Í≥† ÌëúÏ§ÄÌé∏Ï∞®Í∞Ä 1Ïù∏ ÌëúÏ§Ä Ï†ïÍ∑ú Î∂ÑÌè¨ Îç∞Ïù¥ÌÑ∞Î•º 1000Í∞ú ÏÉùÏÑ±Ìï©ÎãàÎã§.

```python
import numpy as np
import matplotlib.pyplot as plt

# Îç∞Ïù¥ÌÑ∞ ÏÉùÏÑ± (ÌèâÍ∑† 0, ÌëúÏ§ÄÌé∏Ï∞® 1Ïù∏ ÎÇúÏàò 1000Í∞ú)
data = np.random.normal(loc=0, scale=1, size=1000)
```

### 2. Í∏∞Ïà† ÌÜµÍ≥ÑÎüâ Í≥ÑÏÇ∞
*   **ÌèâÍ∑† (Mean)**: Îç∞Ïù¥ÌÑ∞Ïùò Î¨¥Í≤å Ï§ëÏã¨
*   **ÌëúÏ§ÄÌé∏Ï∞® (Std)**: Îç∞Ïù¥ÌÑ∞Í∞Ä ÌçºÏ†∏ ÏûàÎäî Ï†ïÎèÑ
*   **Ï§ëÏïôÍ∞í (Median)**: ÏàúÏÑúÎåÄÎ°ú Ï§Ñ ÏÑ∏Ïõ†ÏùÑ Îïå Îî± Ï§ëÍ∞ÑÏóê ÏûàÎäî Í∞í (Ïù¥ÏÉÅÏπòÏóê ÏòÅÌñ• Ï†ÅÏùå)

```python
print(f"ÌèâÍ∑†(Mean): {np.mean(data):.4f}")
print(f"ÌëúÏ§ÄÌé∏Ï∞®(Std): {np.std(data):.4f}")
print(f"Ï§ëÏïôÍ∞í(Median): {np.median(data):.4f}")
```

### 3. Îç∞Ïù¥ÌÑ∞ ÏãúÍ∞ÅÌôî (Histogram)
Îç∞Ïù¥ÌÑ∞Ïùò Î∂ÑÌè¨ Î™®ÏñëÏùÑ ÌûàÏä§ÌÜ†Í∑∏Îû®ÏúºÎ°ú ÌôïÏù∏Ìï©ÎãàÎã§. Ï¢Ö Î™®Ïñë(Bell Curve)Ïù¥ ÎÇòÏò§ÎäîÏßÄ ÌôïÏù∏Ìï¥Î≥¥ÏÑ∏Ïöî.

```python
plt.figure(figsize=(8, 5))
plt.hist(data, bins=30, alpha=0.7, color='skyblue', edgecolor='black')
plt.title("Normal Distribution (Mean=0, Std=1)")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```
