ë°ì´í„° ë¶„ì„ì—ì„œ ê°€ì¥ ì¤‘ìš”í•œ ë‹¨ê³„ì¸ **ê²°ì¸¡ì¹˜(Missing Value)** ì²˜ë¦¬ì™€ **ì´ìƒì¹˜(Outlier)** ì œê±° ë°©ë²•ì„ ì‹¤ìŠµí•©ë‹ˆë‹¤.

```python
import pandas as pd
import numpy as np

# ê°€ìƒì˜ íƒ€ì´íƒ€ë‹‰ ë°ì´í„° ìƒì„± (ê²°ì¸¡ì¹˜ì™€ ì´ìƒì¹˜ í¬í•¨)
data = {
    'Name': ['Jack', 'Rose', 'John', 'Doe'],
    'Age': [20, 17, np.nan, 150],  # NaN: ê²°ì¸¡ì¹˜, 150: ì´ìƒì¹˜
    'Fare': [7.25, 71.28, 8.05, np.nan]
}
df = pd.DataFrame(data)

print("--- ì›ë³¸ ë°ì´í„° (Original) ---")
print(df)
```

### 1. ê²°ì¸¡ì¹˜(NaN) ì²˜ë¦¬
*   **Age**: í‰ê· ê°’ìœ¼ë¡œ ì±„ìš°ê¸° (`fillna(mean)`)
*   **Fare**: 0ìœ¼ë¡œ ì±„ìš°ê¸° (`fillna(0)`)

```python
# Age ê²°ì¸¡ì¹˜ë¥¼ í‰ê· ìœ¼ë¡œ ëŒ€ì²´
mean_age = df['Age'].mean()
df['Age'] = df['Age'].fillna(mean_age)

# Fare ê²°ì¸¡ì¹˜ë¥¼ 0ìœ¼ë¡œ ëŒ€ì²´
df['Fare'] = df['Fare'].fillna(0)

print(f"í‰ê·  ë‚˜ì´: {mean_age:.2f}")
print("\n--- ê²°ì¸¡ì¹˜ ì²˜ë¦¬ í›„ ---")
print(df)
```

### 2. ì´ìƒì¹˜(Outlier) ì²˜ë¦¬
ìƒì‹ì ìœ¼ë¡œ ë¶ˆê°€ëŠ¥í•œ ë°ì´í„°(ì˜ˆ: ë‚˜ì´ê°€ 100ì„¸ ì´ˆê³¼)ë¥¼ ì œê±°í•©ë‹ˆë‹¤.

```python
# ë‚˜ì´ê°€ 100ì‚´ ì´í•˜ì¸ ë°ì´í„°ë§Œ ë‚¨ê¹€
df = df[df['Age'] <= 100]

print("\n--- ì´ìƒì¹˜ ì œê±° í›„ (Final) ---")
print(df)
```

### Practice Challenge
1.  `dropna()`ë¥¼ ì‚¬ìš©í•˜ë©´ ì–´ë–¤ ê²°ê³¼ê°€ ë‚˜ì˜¤ëŠ”ì§€ í…ŒìŠ¤íŠ¸í•´ë³´ì„¸ìš”.
2.  `Fare`ì˜ ì´ìƒì¹˜ëŠ” ì–´ë–»ê²Œ íŒë‹¨í•˜ê³  ì²˜ë¦¬í•  ìˆ˜ ìˆì„ê¹Œìš”?

```python
df2 = pd.DataFrame(data)

df2 = df2.dropna()
print(df2)
```

```python
df3 = pd.DataFrame(data)
# 1. IQR(Interquartile Range) ë°©ì‹ì„ ì´ìš©í•œ ì´ìƒì¹˜ íŒë‹¨
Q1 = df3['Fare'].quantile(0.25)
Q3 = df3['Fare'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# 2. ì´ìƒì¹˜ í™•ì¸
outliers = df3[(df3['Fare'] < lower_bound) | (df3['Fare'] > upper_bound)]
print(f"Detected outliers: {len(outliers)}")

# 3. ì´ìƒì¹˜ ì²˜ë¦¬ (ìƒí•œê°’/í•˜í•œê°’ìœ¼ë¡œ ìº¡í•‘)
df3['Fare'] = df3['Fare'].clip(lower=lower_bound, upper=upper_bound)
```
