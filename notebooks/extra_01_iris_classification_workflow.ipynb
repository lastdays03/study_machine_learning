## ëª©í‘œ
ë¨¸ì‹ ëŸ¬ë‹ì˜ ê°€ì¥ ê¸°ë³¸ì´ ë˜ëŠ” **'ë¶„ë¥˜(Classification)'** í”„ë¡œì„¸ìŠ¤ë¥¼ ì²˜ìŒë¶€í„° ëê¹Œì§€ ì‹¤ìŠµí•©ë‹ˆë‹¤.
ë°ì´í„°ë¥¼ ë¡œë“œí•˜ê³ , ëª¨ë¸ì„ í•™ìŠµì‹œí‚¤ê³ , ìƒˆë¡œìš´ ë°ì´í„°ì˜ ì •ë‹µì„ ì˜ˆì¸¡í•˜ëŠ” ì „ì²´ íë¦„(Workflow)ì„ ìµíˆëŠ” ê²ƒì´ í•µì‹¬ì…ë‹ˆë‹¤.

### í”„ë¡œì„¸ìŠ¤ (Workflow)
1.  **ì¤€ë¹„ (Setup)**: ë¼ì´ë¸ŒëŸ¬ë¦¬ ë° ë°ì´í„° ë¡œë“œ.
2.  **ë¶„ë¦¬ (Split)**: í•™ìŠµìš© ë¬¸ì œì§‘ê³¼ ì‹œí—˜ìš© ë¬¸ì œì§‘ ë‚˜ëˆ„ê¸°.
3.  **í•™ìŠµ (Train)**: ëª¨ë¸(í•™ìƒ)ì—ê²Œ í•™ìŠµìš© ë¬¸ì œì§‘ìœ¼ë¡œ ê³µë¶€ì‹œí‚¤ê¸°.
4.  **ì˜ˆì¸¡ (Predict)**: ì‹œí—˜ìš© ë¬¸ì œì§‘ í’€ê²Œ í•˜ê¸°.
5.  **í‰ê°€ (Evaluate)**: ì±„ì í•˜ê¸° (ì •í™•ë„).

```python
import pandas as pd
import sklearn
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

print(f"scikit-learn version: {sklearn.__version__}")
```

## 1. ë°ì´í„° ë¡œë“œ ë° í™•ì¸
**Why**: ë¨¸ì‹ ëŸ¬ë‹ì˜ ì¬ë£ŒëŠ” ë°ì´í„°ì…ë‹ˆë‹¤. ë°ì´í„°ê°€ ì–´ë–»ê²Œ ìƒê²¼ëŠ”ì§€(Feature), ë¬´ì—‡ì„ ë§ì¶°ì•¼ í•˜ëŠ”ì§€(Label) í™•ì¸í•´ì•¼ í•©ë‹ˆë‹¤.

```python
# ë¶“ê½ƒ ë°ì´í„° ì„¸íŠ¸ ë¡œë”©
iris = load_iris()
iris_data = iris.data
iris_label = iris.target

print('iris targetê°’:', iris_label[:10])
print('iris targetëª…:', iris.target_names)

# ë³´ê¸° í¸í•˜ê²Œ DataFrameìœ¼ë¡œ ë³€í™˜
df_iris = pd.DataFrame(data=iris_data, columns=iris.feature_names)
df_iris['label'] = iris.target
df_iris.head(3)
```

## 2. í•™ìŠµ/í…ŒìŠ¤íŠ¸ ë°ì´í„° ë¶„ë¦¬ (Train/Test Split)
**Why**: 
*   ëª¨ë“  ë°ì´í„°ë¡œ ê³µë¶€í•˜ê³  ê·¸ ë°ì´í„°ë¡œ ì‹œí—˜ì„ ì¹˜ë©´ 100ì ì´ ë‚˜ì˜¤ê² ì§€ë§Œ, ì‹¤ë ¥ì„ ê²€ì¦í•  ìˆ˜ ì—†ìŠµë‹ˆë‹¤. (ì´ê²ƒì„ **ê³¼ì í•©, Overfitting**ì´ë¼ê³  í•©ë‹ˆë‹¤.)
*   ë”°ë¼ì„œ ê³µë¶€í•  ë°ì´í„°(Train)ì™€ ê²€ì¦í•  ë°ì´í„°(Test)ë¥¼ ë¶„ë¦¬í•´ì•¼ í•©ë‹ˆë‹¤.

```python
# test_size=0.2 : ì „ì²´ ë°ì´í„°ì˜ 20%ë¥¼ í…ŒìŠ¤íŠ¸ìš©ìœ¼ë¡œ ë–¼ì–´ë†“ìŒ
# random_state=11 : ë§¤ë²ˆ ì‹¤í–‰í•  ë•Œë§ˆë‹¤ ë˜‘ê°™ì´ ì„ì´ë„ë¡ ê³ ì • (ì‹¤ìŠµìš©)
X_train, X_test, y_train, y_test = train_test_split(iris_data, iris_label, test_size=0.2, random_state=11)

print(f"í•™ìŠµì…‹ í¬ê¸°: {X_train.shape}, í…ŒìŠ¤íŠ¸ì…‹ í¬ê¸°: {X_test.shape}")
```

## 3. ëª¨ë¸ í•™ìŠµ (Training)
**Why**: **ì˜ì‚¬ê²°ì •íŠ¸ë¦¬(Decision Tree)** ì•Œê³ ë¦¬ì¦˜ì„ ì‚¬ìš©í•©ë‹ˆë‹¤. ìŠ¤ë¬´ê³ ê°œì™€ ë¹„ìŠ·í•˜ê²Œ "ê½ƒì ê¸¸ì´ê°€ Xë³´ë‹¤ ê¸´ê°€?" ê°™ì€ ì§ˆë¬¸ì„ ë°˜ë³µí•˜ë©° ì •ë‹µì„ ì°¾ì•„ê°€ëŠ” ì§ê´€ì ì¸ ëª¨ë¸ì…ë‹ˆë‹¤.

```python
# ê°ì²´ ìƒì„±
dt_clf = DecisionTreeClassifier(random_state=11)

# í•™ìŠµ ìˆ˜í–‰ (fit) -> ëª¨ë¸ì´ íŒ¨í„´ì„ í•™ìŠµí•¨
dt_clf.fit(X_train, y_train)
```

## 4. ì˜ˆì¸¡ ìˆ˜í–‰ (Prediction)
**Why**: í•™ìŠµë˜ì§€ ì•Šì€ ìƒˆë¡œìš´ ë°ì´í„°(Test Set)ì— ëŒ€í•´ ëª¨ë¸ì´ ìŠ¤ìŠ¤ë¡œ íŒë‹¨ì„ ë‚´ë¦¬ê²Œ í•©ë‹ˆë‹¤.

```python
# í•™ìŠµì— ì‚¬ìš©í•˜ì§€ ì•Šì€ í…ŒìŠ¤íŠ¸ ë°ì´í„°ë¡œ ì˜ˆì¸¡
pred = dt_clf.predict(X_test)
print(f"ì˜ˆì¸¡ê°’(ìƒìœ„ 5ê°œ): {pred[:5]}")
```

## 5. ì˜ˆì¸¡ ì •í™•ë„ í‰ê°€ (Evaluation)
**Why**: ëª¨ë¸ì´ ì–¼ë§ˆë‚˜ ì˜ ë§ì·„ëŠ”ì§€ ì ìˆ˜ë¥¼ ë§¤ê²¨ì•¼ ì„±ëŠ¥ì„ ì•Œ ìˆ˜ ìˆìŠµë‹ˆë‹¤. ì—¬ê¸°ì„œëŠ” ê°€ì¥ ì§ê´€ì ì¸ **ì •í™•ë„(Accuracy)**ë¥¼ ì‚¬ìš©í•©ë‹ˆë‹¤.

```python
score = accuracy_score(y_test, pred)
print(f"ì˜ˆì¸¡ ì •í™•ë„: {score:.4f}")
```

*   **ë°ì´í„° ë¶„ë¦¬ì˜ ì¤‘ìš”ì„±**: í•™ìŠµí•œ ë¬¸ì œë¡œ ì‹œí—˜ì„ ì¹˜ë©´ ì‹¤ë ¥ì„ ì•Œ ìˆ˜ ì—†ë“¯ì´, ë¨¸ì‹ ëŸ¬ë‹ì—ì„œë„ **Train Set**ê³¼ **Test Set**ì˜ ë¶„ë¦¬ëŠ” í•„ìˆ˜ì ì…ë‹ˆë‹¤.
