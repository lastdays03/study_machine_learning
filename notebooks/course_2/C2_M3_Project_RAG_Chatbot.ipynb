```python
# ğŸ”§ Setup & Imports
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

## 1. ğŸ“š RAG ì§€ì‹ ë² ì´ìŠ¤ êµ¬ì¶•
ë¬¸ì„œë¥¼ ìª¼ê°œê³ (Chunking), ë²¡í„°ë¡œ ë³€í™˜(Embedding)í•˜ì—¬ ì €ì¥í•©ë‹ˆë‹¤.


```python
docs = [
    'MLflow is a tool for managing the ML lifecycle.',
    'FastAPI is a web framework for building APIs.',
    'Docker is a platform for containerization.'
]
vectorizer = TfidfVectorizer()
doc_embeddings = vectorizer.fit_transform(docs)
print(f'Knowledge Base Created with {len(docs)} documents.')
```

## 2. ğŸ”¦ ê²€ìƒ‰(Retrieval) ê³¼ì •
ì§ˆë¬¸ -> ë²¡í„°í™” -> ì½”ì‚¬ì¸ ìœ ì‚¬ë„ ê³„ì‚° -> ê°€ì¥ ê´€ë ¨ì„± ë†’ì€ ë¬¸ì„œ ì¶”ì¶œ


```python
query = 'What tool helps with ML lifecycle?'
query_vec = vectorizer.transform([query])

scores = cosine_similarity(query_vec, doc_embeddings).flatten()
best_doc_idx = scores.argmax()

print(f'Query: {query}')
print(f'Retrieved: {docs[best_doc_idx]} (Score: {scores[best_doc_idx]:.2f})')
```
