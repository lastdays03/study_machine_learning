```python
# üîß Setup & Imports
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

## 1. Knowledge Base (Chunks)


```python
docs = [
    'MLflow is an open source platform for managing the ML lifecycle.',
    'FastAPI is a modern web framework for building APIs with Python.',
    'Docker uses containerization to package applications.',
    'Transformers provided by Hugging Face are state-of-the-art models.'
]
```

## 2. Vector DB (Simulated with TF-IDF/Cosine)


```python

vectorizer = TfidfVectorizer()
embeddings = vectorizer.fit_transform(docs)
print('Embeddings Shape:', embeddings.shape)
```

## 3. Retrieval System


```python
query = 'How do I deploy models?'
query_vec = vectorizer.transform([query])

scores = cosine_similarity(query_vec, embeddings).flatten()
best_idx = scores.argmax()

print('Query:', query)
print('Best Doc:', docs[best_idx])
print('Score:', scores[best_idx])
```
