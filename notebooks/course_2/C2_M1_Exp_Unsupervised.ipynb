```python
# ðŸ”§ Setup & Imports
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import numpy as np
```

## 1. Data Generation (Synthetic)
We generate isotropic Gaussian blobs for clustering.


```python

# Generate Data
X, y = make_blobs(n_samples=500, centers=4, cluster_std=1.0, random_state=42)
print('Data Shape:', X.shape)
```

## 2. K-Means Clustering
Finding 4 clusters.


```python
kmeans = KMeans(n_clusters=4, random_state=42)
y_pred = kmeans.fit_predict(X)

# Visualize
plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], c=y_pred, cmap='viridis', marker='.')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], c='red', s=100, marker='x', label='Centroids')
plt.title('K-Means Clustering Result')
plt.legend()
plt.show()
```

## 3. PCA Dimension Reduction
Reducing to 2 components (though our data is already 2D, this demonstrates the API).


```python
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)
print('Explained Variance Ratio:', pca.explained_variance_ratio_)
```
