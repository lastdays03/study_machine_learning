```python
# üîß Setup & Imports
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import numpy as np
```

## 1. üîç Ïù¥Î°†Ï†Å Î∞∞Í≤Ω: ÌÅ¥Îü¨Ïä§ÌÑ∞ÎßÅ (Clustering)
> **ÎπÑÏßÄÎèÑ ÌïôÏäµ(Unsupervised Learning)**ÏùÄ Ï†ïÎãµ(Label)Ïù¥ ÏóÜÎäî Îç∞Ïù¥ÌÑ∞ÏóêÏÑú Ïà®Í≤®ÏßÑ Ìå®ÌÑ¥ÏùÑ Ï∞æÏïÑÎÉÖÎãàÎã§.

### Ïôú K-MeansÏù∏Í∞Ä?
*   **Î™©Ìëú**: Îç∞Ïù¥ÌÑ∞Î•º $K$Í∞úÏùò Í∑∏Î£πÏúºÎ°ú Î¨∂Îäî Í≤É.
*   **ÏõêÎ¶¨**: Ï§ëÏã¨Ï†ê(Centroid)ÏùÑ Î∞òÎ≥µÏ†ÅÏúºÎ°ú ÏóÖÎç∞Ïù¥Ìä∏ÌïòÏó¨ Í∑∏Î£π ÎÇ¥ Î∂ÑÏÇ∞(Variance)ÏùÑ ÏµúÏÜåÌôî.
*   **ÌïúÍ≥Ñ**: $K$Î•º ÎØ∏Î¶¨ ÏßÄÏ†ïÌï¥Ïïº Ìï®.


## 2. üìä Îç∞Ïù¥ÌÑ∞ ÏÉùÏÑ± Î∞è ÏãúÍ∞ÅÌôî
Ïã§ÏäµÏùÑ ÏúÑÌï¥ 4Í∞úÏùò ÌÅ¥Îü¨Ïä§ÌÑ∞Î•º Í∞ÄÏßÑ Í∞ÄÏÉÅ Îç∞Ïù¥ÌÑ∞Î•º ÏÉùÏÑ±Ìï©ÎãàÎã§.


```python
X, y = make_blobs(n_samples=500, centers=4, cluster_std=1.0, random_state=42)

# Raw Data ÏãúÍ∞ÅÌôî
plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], c='lightgray', marker='.', s=50, label='Data Points')
plt.title('Synthetic Data for Clustering')
plt.legend()
plt.show()
```

## 3. üß† Î™®Îç∏ ÌïôÏäµ: K-Means
ÏÑ§Ï†ïÌïú $K=4$Î°ú Î™®Îç∏ÏùÑ ÌïôÏäµÏãúÌÇµÎãàÎã§.


```python
kmeans = KMeans(n_clusters=4, random_state=42)
y_pred = kmeans.fit_predict(X)
print('Inertia (WSSE):', kmeans.inertia_)
```

## 4. üìà Í≤∞Í≥º ÏãúÍ∞ÅÌôî
Î™®Îç∏Ïù¥ ÏòàÏ∏°Ìïú ÌÅ¥Îü¨Ïä§ÌÑ∞ÏôÄ Í≥ÑÏÇ∞Îêú Ï§ëÏã¨Ï†ê(Îπ®Í∞Ñ X)ÏùÑ ÏãúÍ∞ÅÌôîÌï©ÎãàÎã§.


```python
plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], c=y_pred, cmap='viridis', marker='.', s=50, alpha=0.6)
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], c='red', s=200, marker='x', label='Centroids')
plt.title('K-Means Clustering Result (K=4)')
plt.legend()
plt.show()
```

## 5. üî¨ Ïã§Ìóò: Elbow Method
ÏµúÏ†ÅÏùò KÎäî Ïñ¥ÎñªÍ≤å Ï∞æÏùÑÍπåÏöî? **Elbow Method**Î•º ÌÜµÌï¥ Í≤ÄÏ¶ùÌï¥Î¥ÖÏãúÎã§.
KÎ•º 1Î∂ÄÌÑ∞ 9ÍπåÏßÄ ÎäòÎ†§Í∞ÄÎ©∞ InertiaÍ∞Ä Í∏âÍ≤©Ìûà Í∫æÏù¥Îäî ÏßÄÏ†êÏùÑ Ï∞æÏäµÎãàÎã§.


```python
inertias = []
k_range = range(1, 10)
for k in k_range:
    km = KMeans(n_clusters=k, random_state=42)
    km.fit(X)
    inertias.append(km.inertia_)

plt.figure(figsize=(8, 4))
plt.plot(k_range, inertias, 'bo-')
plt.xlabel('Number of Clusters K')
plt.ylabel('Inertia')
plt.title('Elbow Method For Optimal k')
plt.grid(True)
plt.show()
```
