```python
# üîß Setup & Imports
import mlflow
import os
```

## 1. MLflow Tracking Setup


```python

# Set local tracking URI
mlflow.set_tracking_uri('file:./mlruns')
mlflow.set_experiment('My_First_Experiment')
```

## 2. Run Experiment


```python
with mlflow.start_run():
    # Log Params
    param1 = 5
    mlflow.log_param('param1', param1)
    
    # Log Metrics
    metric = 0.85
    mlflow.log_metric('accuracy', metric)
    
    # Log Artifacts (Dummy file)
    with open('dummy.txt', 'w') as f:
        f.write('Hello MLflow!')
    mlflow.log_artifact('dummy.txt')
    
    print('Run ID:', mlflow.active_run().info.run_id)
```
