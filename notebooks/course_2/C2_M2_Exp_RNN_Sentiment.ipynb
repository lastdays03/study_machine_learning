```python
# üîß Setup & Imports
import torch
import torch.nn as nn
```

## 1. Tokenization & Embedding (Mock)
Simulating words as integers.


```python

vocab_size = 100
embed_dim = 16
hidden_dim = 32

embedding = nn.Embedding(vocab_size, embed_dim)

# Input: Batch of 2 sentences, length 5
input_seq = torch.LongTensor([[1, 5, 2, 9, 0], [2, 8, 3, 1, 0]])
embedded = embedding(input_seq)
print('Embedded Shape:', embedded.shape) # (Batch, Seq, Embed)
```

## 2. LSTM Model


```python
lstm = nn.LSTM(input_size=embed_dim, hidden_size=hidden_dim, batch_first=True)
output, (hidden, cell) = lstm(embedded)
print('LSTM Output Shape:', output.shape)
print('Final Hidden State Shape:', hidden.shape)
```
