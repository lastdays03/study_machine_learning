```python
vocab_size = 100
embed_dim = 16
embedding = nn.Embedding(vocab_size, embed_dim)

input_indices = torch.LongTensor([[1, 5, 2], [9, 4, 1]]) # Batch=2, Seq=3
embedded = embedding(input_indices)
print(f'Embedding Output Shape: {embedded.shape} (Batch, Seq, Feature)')
```

## 2. ğŸ” LSTM êµ¬ì¡° í™•ì¸

```python
lstm = nn.LSTM(input_size=embed_dim, hidden_size=8, batch_first=True)
output, (hidden, cell) = lstm(embedded)

print('LSTM Output sequence:', output.shape)
print('Final Hidden State:', hidden.shape)
```
