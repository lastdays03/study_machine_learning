```python
# üîß Setup & Imports
import torch
import torch.nn as nn
```

## 1. Define CNN Architecture
Convolution -> ReLU -> Pool -> Linear


```python

class SimpleCNN(nn.Module):
    def __init__(self):
        super(SimpleCNN, self).__init__()
        # Input: 1 channel (grayscale), Output: 6 channels, Kernel: 3x3
        self.conv1 = nn.Conv2d(1, 6, 3)
        self.pool = nn.MaxPool2d(2, 2)
        self.conv2 = nn.Conv2d(6, 16, 3)
        # Fully Connected
        self.fc1 = nn.Linear(16 * 6 * 6, 10)
    
    def forward(self, x):
        # Input shape: (Batch, 1, 28, 28) assumed for MNIST-like
        x = self.pool(torch.relu(self.conv1(x)))
        x = self.pool(torch.relu(self.conv2(x)))
        x = torch.flatten(x, 1)
        x = self.fc1(x)
        return x

model = SimpleCNN()
print(model)
```

## 2. Dummy Forward Pass
Checking dimension flow.


```python
# Create a random image batch (Batch=1, Channel=1, Height=32, Width=32)
dummy_input = torch.randn(1, 1, 32, 32)
try:
    output = model(dummy_input)
    print('Output Shape:', output.shape)
except Exception as e:
    print('Error (likely mismatch dimension):', e)
```
