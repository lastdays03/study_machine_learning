```python
import torch.nn as nn
# üîß Setup & Imports
import matplotlib.pyplot as plt
import seaborn as sns
import torch
import torch.nn.functional as F
```

## 1. üëÅÔ∏è Ïñ¥ÌÖêÏÖò(Attention) Î©îÏª§ÎãàÏ¶ò
**Í≥µÏãù**: $Attention(Q, K, V) = softmax(\frac{QK^T}{\sqrt{d_k}})V$

QueryÏôÄ KeyÏùò Ïú†ÏÇ¨ÎèÑÎ•º Í≥ÑÏÇ∞ÌïòÏó¨, Ïñ¥Îñ§ Îã®Ïñ¥(Value)Ïóê ÏßëÏ§ëÌï†ÏßÄ Í≤∞Ï†ïÌï©ÎãàÎã§.


```python
def self_attention(query, key, value):
    d_k = query.size(-1)
    scores = torch.matmul(query, key.transpose(-2, -1)) / torch.sqrt(torch.tensor(d_k, dtype=torch.float))
    weights = F.softmax(scores, dim=-1)
    return torch.matmul(weights, value), weights

d_model = 16
seq_len = 10
x = torch.randn(1, seq_len, d_model)
out, attn_weights = self_attention(x, x, x)
print('Attention Computed.')
```

## 2. üî• Ïñ¥ÌÖêÏÖò Í∞ÄÏ§ëÏπò ÏãúÍ∞ÅÌôî
Ïñ¥Îñ§ ÌÜ†ÌÅ∞ÎÅºÎ¶¨ Í∞ïÌïòÍ≤å Ïó∞Í≤∞ÎêòÏóàÎäîÏßÄ ÌûàÌä∏ÎßµÏúºÎ°ú ÌôïÏù∏Ìï©ÎãàÎã§.


```python
plt.figure(figsize=(8, 6))
sns.heatmap(attn_weights[0].numpy(), cmap='Reds', square=True)
plt.title('Self-Attention Map (Random Data)')
plt.xlabel('Key Position')
plt.ylabel('Query Position')
plt.show()
```
