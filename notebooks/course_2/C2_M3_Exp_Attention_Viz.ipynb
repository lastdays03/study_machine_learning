```python
# üîß Setup & Imports
import matplotlib.pyplot as plt
import seaborn as sns
import torch
import torch.nn.functional as F
```

## 1. Scaled Dot-Product Attention from Scratch


```python

def self_attention(query, key, value):
    d_k = query.size(-1)
    scores = torch.matmul(query, key.transpose(-2, -1)) / torch.sqrt(torch.tensor(d_k, dtype=torch.float))
    weights = F.softmax(scores, dim=-1)
    return torch.matmul(weights, value), weights

# Random Q, K, V
seq_len = 10
d_model = 64
x = torch.randn(1, seq_len, d_model)

output, attn_weights = self_attention(x, x, x)
print('Attention Output:', output.shape)
```

## 2. Visualize Attention Weights


```python
plt.figure(figsize=(8, 6))
sns.heatmap(attn_weights[0].detach().numpy(), cmap='Reds')
plt.title('Self-Attention Map (Random)')
plt.xlabel('Key Position')
plt.ylabel('Query Position')
plt.show()
```
